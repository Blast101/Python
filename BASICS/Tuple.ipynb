# Python Tuple Methods & Operations Reference

A comprehensive guide to Python tuples with all methods, operations, and examples.

**Note:** Tuples are **immutable** - once created, they cannot be modified!

## Creating Tuples

**Description:** Different ways to create tuples in Python.

```python
# Method 1: Using parentheses
fruits = ('apple', 'banana', 'orange')
print(f"Fruits: {fruits}")

# Method 2: Without parentheses (tuple packing)
numbers = 1, 2, 3, 4
print(f"Numbers: {numbers}")

# Method 3: Single element tuple (note the comma!)
single = (5,)
print(f"Single element: {single}")
print(f"Type: {type(single)}")

# Method 4: Empty tuple
empty = ()
print(f"Empty: {empty}")

# Method 5: Using tuple() constructor
from_list = tuple([1, 2, 3])
print(f"From list: {from_list}")

from_string = tuple('hello')
print(f"From string: {from_string}")
```

## Accessing Tuple Elements

**Description:** How to access individual elements and slices from tuples.

```python
# Creating a sample tuple
fruits = ('apple', 'banana', 'orange', 'mango', 'grape')

# Positive indexing (starts from 0)
print(f"First element: {fruits[0]}")   # Output: apple
print(f"Third element: {fruits[2]}")   # Output: orange

# Negative indexing (starts from -1)
print(f"Last element: {fruits[-1]}")    # Output: grape
print(f"Second last: {fruits[-2]}")     # Output: mango

# Slicing [start:stop:step]
print(f"First three: {fruits[0:3]}")     # Output: ('apple', 'banana', 'orange')
print(f"From index 2: {fruits[2:]}")     # Output: ('orange', 'mango', 'grape')
print(f"Up to index 3: {fruits[:3]}")    # Output: ('apple', 'banana', 'orange')
print(f"Every second: {fruits[::2]}")    # Output: ('apple', 'orange', 'grape')
print(f"Reversed: {fruits[::-1]}")       # Output: ('grape', 'mango', 'orange', 'banana', 'apple')
```

## 1. `count()` Method

**Description:** Returns the number of times a specified element appears in the tuple.

**Syntax:** `tuple.count(element)`

**Returns:** Integer (count of occurrences)

```python
# Example 1: Count occurrences in fruits
fruits = ('apple', 'banana', 'orange', 'banana', 'banana')
count = fruits.count('banana')
print(f"Count of banana: {count}")  # Output: Count of banana: 3

# Example 2: Count in numbers
numbers = (1, 2, 3, 2, 4, 2, 5)
count = numbers.count(2)
print(f"Count of 2: {count}")  # Output: Count of 2: 3

# Example 3: Element not in tuple returns 0
count = fruits.count('grape')
print(f"Count of grape: {count}")  # Output: Count of grape: 0

# Example 4: Count in nested tuples
nested = (1, 2, (3, 4), 2, (3, 4))
count = nested.count((3, 4))
print(f"Count of (3, 4): {count}")  # Output: Count of (3, 4): 2
```

## 2. `index()` Method

**Description:** Returns the index of the first occurrence of a specified element. Optional start and end parameters can limit the search range.

**Syntax:** `tuple.index(element, start=0, end=len(tuple))`

**Returns:** Integer (index of the element)

**Raises:** ValueError if element is not found

```python
# Example 1: Find index of element
fruits = ('apple', 'banana', 'orange', 'banana')
idx = fruits.index('banana')
print(f"Index of banana: {idx}")  # Output: Index of banana: 1

# Example 2: Find with duplicates
numbers = (10, 20, 30, 20, 40)
idx = numbers.index(20)
print(f"Index of 20: {idx}")  # Output: Index of 20: 1

# Example 3: Search in a specific range
idx = numbers.index(20, 2)  # Start from index 2
print(f"Index of 20 (from index 2): {idx}")  # Output: Index of 20 (from index 2): 3

# Example 4: Search with start and end
idx = numbers.index(20, 1, 4)  # Search between index 1 and 4
print(f"Index of 20 (between 1 and 4): {idx}")  # Output: Index of 20 (between 1 and 4): 1

# Example 5: Error when not found (commented to avoid error)
# fruits.index('grape')  # ValueError: tuple.index(x): x not in tuple
```

## Tuple Packing and Unpacking

**Description:** Packing multiple values into a tuple and unpacking tuple values into variables.

```python
# Tuple Packing
coordinates = 10, 20, 30
print(f"Packed tuple: {coordinates}")

# Tuple Unpacking
x, y, z = coordinates
print(f"x={x}, y={y}, z={z}")

# Unpacking with * operator (extended unpacking)
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(f"First: {first}")    # Output: First: 1
print(f"Middle: {middle}")  # Output: Middle: [2, 3, 4]
print(f"Last: {last}")      # Output: Last: 5

# Swapping values using tuples
a, b = 5, 10
print(f"Before swap: a={a}, b={b}")
a, b = b, a
print(f"After swap: a={a}, b={b}")

# Function returning multiple values (as tuple)
def get_min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = get_min_max([1, 5, 3, 9, 2])
print(f"Min: {minimum}, Max: {maximum}")
```

## Tuple Concatenation and Repetition

**Description:** Combining tuples and repeating tuple elements.

```python
# Concatenation using + operator
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined = tuple1 + tuple2
print(f"Combined: {combined}")  # Output: Combined: (1, 2, 3, 4, 5, 6)

# Repetition using * operator
original = (1, 2)
repeated = original * 3
print(f"Repeated: {repeated}")  # Output: Repeated: (1, 2, 1, 2, 1, 2)

# Multiple concatenations
fruits = ('apple',)
vegetables = ('carrot',)
grains = ('rice',)
food = fruits + vegetables + grains
print(f"Food: {food}")  # Output: Food: ('apple', 'carrot', 'rice')
```

## Membership Testing

**Description:** Check if an element exists in a tuple using `in` and `not in` operators.

```python
# Create a sample tuple
fruits = ('apple', 'banana', 'orange', 'mango')

# Check membership with 'in'
print(f"'apple' in fruits: {'apple' in fruits}")      # Output: True
print(f"'grape' in fruits: {'grape' in fruits}")      # Output: False

# Check with 'not in'
print(f"'grape' not in fruits: {'grape' not in fruits}")  # Output: True

# Using in conditional statements
if 'banana' in fruits:
    print("Banana is available!")

# With numbers
numbers = (1, 2, 3, 4, 5)
search = 3
result = "Found" if search in numbers else "Not found"
print(f"{search} is {result}")
```

## Built-in Functions with Tuples

**Description:** Common built-in functions that work with tuples.

```python
# Sample tuple
numbers = (5, 2, 8, 1, 9, 3)

# len() - Get length of tuple
print(f"Length: {len(numbers)}")  # Output: Length: 6

# min() - Get minimum value
print(f"Minimum: {min(numbers)}")  # Output: Minimum: 1

# max() - Get maximum value
print(f"Maximum: {max(numbers)}")  # Output: Maximum: 9

# sum() - Get sum of all elements
print(f"Sum: {sum(numbers)}")  # Output: Sum: 28

# sorted() - Return sorted list (doesn't modify original)
sorted_list = sorted(numbers)
print(f"Sorted (as list): {sorted_list}")  # Output: Sorted (as list): [1, 2, 3, 5, 8, 9]
print(f"Original tuple: {numbers}")  # Output: Original tuple: (5, 2, 8, 1, 9, 3)

# sorted() with tuple output
sorted_tuple = tuple(sorted(numbers))
print(f"Sorted (as tuple): {sorted_tuple}")  # Output: Sorted (as tuple): (1, 2, 3, 5, 8, 9)

# reversed() - Return reversed iterator
reversed_tuple = tuple(reversed(numbers))
print(f"Reversed: {reversed_tuple}")  # Output: Reversed: (3, 9, 1, 8, 2, 5)

# any() - Check if any element is True
bool_tuple = (False, False, True, False)
print(f"Any True: {any(bool_tuple)}")  # Output: Any True: True

# all() - Check if all elements are True
print(f"All True: {all(bool_tuple)}")  # Output: All True: False
```

## Iterating Through Tuples

**Description:** Different ways to loop through tuple elements.

```python
fruits = ('apple', 'banana', 'orange', 'mango')

# Method 1: Simple for loop
print("Method 1: Simple iteration")
for fruit in fruits:
    print(f"  {fruit}")

# Method 2: With index using enumerate()
print("\nMethod 2: With index")
for index, fruit in enumerate(fruits):
    print(f"  {index}: {fruit}")

# Method 3: With custom start index
print("\nMethod 3: Custom start index")
for index, fruit in enumerate(fruits, start=1):
    print(f"  {index}. {fruit}")

# Method 4: Using range and indexing
print("\nMethod 4: Using range")
for i in range(len(fruits)):
    print(f"  fruits[{i}] = {fruits[i]}")

# Method 5: Iterating through multiple tuples with zip()
print("\nMethod 5: Multiple tuples")
colors = ('red', 'yellow', 'orange', 'orange')
for fruit, color in zip(fruits, colors):
    print(f"  {fruit} is {color}")
```

## Nested Tuples

**Description:** Tuples containing other tuples (multi-dimensional tuples).

```python
# Creating nested tuples
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)
print(f"Matrix: {matrix}")

# Accessing nested elements
print(f"First row: {matrix[0]}")        # Output: (1, 2, 3)
print(f"Element [1][2]: {matrix[1][2]}")  # Output: 6

# Nested tuple with mixed data
person = ('John', 25, ('New York', 'USA'))
name, age, (city, country) = person
print(f"Name: {name}, Age: {age}, City: {city}, Country: {country}")

# Iterating through nested tuples
coordinates = ((1, 2), (3, 4), (5, 6))
print("\nCoordinates:")
for x, y in coordinates:
    print(f"  ({x}, {y})")
```

## Tuple Immutability

**Description:** Understanding that tuples cannot be modified after creation.

```python
# Create a tuple
fruits = ('apple', 'banana', 'orange')
print(f"Original: {fruits}")

# These operations will raise errors (commented to avoid errors)
# fruits[0] = 'grape'  # TypeError: 'tuple' object does not support item assignment
# fruits.append('mango')  # AttributeError: 'tuple' object has no attribute 'append'
# del fruits[1]  # TypeError: 'tuple' object doesn't support item deletion

# However, you CAN:

# 1. Create a new tuple
new_fruits = ('grape',) + fruits[1:]
print(f"New tuple: {new_fruits}")

# 2. Convert to list, modify, convert back
fruits_list = list(fruits)
fruits_list[0] = 'grape'
fruits = tuple(fruits_list)
print(f"Modified via list: {fruits}")

# 3. Reassign the variable to a new tuple
fruits = ('kiwi', 'melon', 'peach')
print(f"Reassigned: {fruits}")

# Important: Mutable objects inside tuples CAN be modified
mixed = (1, 2, [3, 4], 5)
print(f"\nBefore: {mixed}")
mixed[2][0] = 99  # Modifying the list inside the tuple
print(f"After: {mixed}")  # The tuple's reference doesn't change, but the list does
```

## Tuple vs List Comparison

**Description:** Key differences between tuples and lists.

```python
import sys

# Create similar tuple and list
my_tuple = (1, 2, 3, 4, 5)
my_list = [1, 2, 3, 4, 5]

# 1. Memory usage
print("Memory Comparison:")
print(f"  Tuple size: {sys.getsizeof(my_tuple)} bytes")
print(f"  List size: {sys.getsizeof(my_list)} bytes")
print(f"  Difference: Tuples use less memory!")

# 2. Performance (iteration speed)
import timeit
print("\nPerformance Comparison:")
tuple_time = timeit.timeit('for i in (1, 2, 3, 4, 5): pass', number=1000000)
list_time = timeit.timeit('for i in [1, 2, 3, 4, 5]: pass', number=1000000)
print(f"  Tuple iteration: {tuple_time:.6f} seconds")
print(f"  List iteration: {list_time:.6f} seconds")
print(f"  Difference: Tuples are slightly faster!")

# 3. Available methods
print("\nMethod Comparison:")
tuple_methods = [m for m in dir(my_tuple) if not m.startswith('_')]
list_methods = [m for m in dir(my_list) if not m.startswith('_')]
print(f"  Tuple methods: {tuple_methods}")
print(f"  List has {len(list_methods)} methods, Tuple has {len(tuple_methods)} methods")
print(f"  Lists have more methods because they're mutable!")
```

## When to Use Tuples

### âœ… Use Tuples When:
- Data should not change (immutability)
- You need a hashable type (for dictionary keys or set elements)
- You want better performance and less memory usage
- Representing fixed collections (coordinates, RGB colors, etc.)
- Returning multiple values from functions

### âŒ Use Lists When:
- You need to modify data (add, remove, change elements)
- You need list-specific methods (append, extend, sort, etc.)
- Data represents a collection that may grow or shrink

## Tips & Best Practices

### ðŸ’¡ Single Element Tuple
Always use a trailing comma for single-element tuples: `(5,)` not `(5)`

### âš ï¸ Immutability
Remember that tuples are immutable. Use lists if you need to modify data.

### ðŸš€ Performance
Tuples are faster and use less memory than lists. Use them when data won't change.

### ðŸ“¦ Unpacking
Use tuple unpacking for cleaner code when working with functions that return multiple values.

### ðŸ”‘ Dictionary Keys
Tuples can be used as dictionary keys (if they contain only immutable elements), lists cannot.
