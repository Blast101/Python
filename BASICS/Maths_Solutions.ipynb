## Converting digits to an Integer and visa versa

```python
numbers = [3,4,5,2]

total = 0
for digit in numbers:
    total = total*10 + digit

print(total)

digits = 0
numbers = []
while total>0:
    reminder = total % 10
    numbers.append(reminder)
    digits = digits * 10 + reminder
    total = total // 10
    
print(numbers)
    
```
