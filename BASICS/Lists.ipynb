# Python List Methods Reference

A comprehensive guide to all Python list methods with examples and descriptions.

## 1. `append()` Method

**Description:** Adds a single element to the end of the list.

**Syntax:** `list.append(element)`

**Returns:** None (modifies the list in-place)

```python
# Example 1: Append to fruits list
fruits = ['apple', 'banana']
fruits.append('orange')
print(fruits)  # Output: ['apple', 'banana', 'orange']

# Example 2: Append to numbers list
numbers = [1, 2, 3]
numbers.append(4)
print(numbers)  # Output: [1, 2, 3, 4]
```

## 2. `extend()` Method

**Description:** Adds all elements from an iterable (list, tuple, string, etc.) to the end of the list.

**Syntax:** `list.extend(iterable)`

**Returns:** None (modifies the list in-place)

```python
# Example 1: Extend with another list
fruits = ['apple', 'banana']
fruits.extend(['orange', 'mango'])
print(fruits)  # Output: ['apple', 'banana', 'orange', 'mango']

# Example 2: Extend with numbers
numbers = [1, 2]
numbers.extend([3, 4, 5])
print(numbers)  # Output: [1, 2, 3, 4, 5]

# Example 3: Extend with a string
letters = ['a', 'b']
letters.extend('cd')
print(letters)  # Output: ['a', 'b', 'c', 'd']
```

## 3. `insert()` Method

**Description:** Inserts an element at a specified position in the list.

**Syntax:** `list.insert(index, element)`

**Returns:** None (modifies the list in-place)

```python
# Example 1: Insert in the middle
fruits = ['apple', 'orange']
fruits.insert(1, 'banana')
print(fruits)  # Output: ['apple', 'banana', 'orange']

# Example 2: Insert at specific index
numbers = [1, 3, 4]
numbers.insert(1, 2)
print(numbers)  # Output: [1, 2, 3, 4]

# Example 3: Insert at the beginning
colors = ['red', 'blue']
colors.insert(0, 'green')
print(colors)  # Output: ['green', 'red', 'blue']
```

## 4. `remove()` Method

**Description:** Removes the first occurrence of a specified element from the list.

**Syntax:** `list.remove(element)`

**Returns:** None (modifies the list in-place)

**Raises:** ValueError if element is not found

```python
# Example 1: Remove first occurrence
fruits = ['apple', 'banana', 'orange', 'banana']
fruits.remove('banana')
print(fruits)  # Output: ['apple', 'orange', 'banana']

# Example 2: Remove from numbers
numbers = [1, 2, 3, 2, 4]
numbers.remove(2)
print(numbers)  # Output: [1, 3, 2, 4]

# Example 3: Error when element not found (commented to avoid error)
# fruits.remove('grape')  # ValueError: list.remove(x): x not in list
```

## 5. `pop()` Method

**Description:** Removes and returns the element at a specified position. If no index is specified, removes and returns the last element.

**Syntax:** `list.pop(index=-1)`

**Returns:** The removed element

**Raises:** IndexError if list is empty or index out of range

```python
# Example 1: Pop last element
fruits = ['apple', 'banana', 'orange']
last = fruits.pop()
print(f"Popped: {last}")    # Output: Popped: orange
print(f"Remaining: {fruits}")  # Output: Remaining: ['apple', 'banana']

# Example 2: Pop at specific index
numbers = [1, 2, 3, 4]
second = numbers.pop(1)
print(f"Popped: {second}")   # Output: Popped: 2
print(f"Remaining: {numbers}")  # Output: Remaining: [1, 3, 4]

# Example 3: Error on empty list (commented to avoid error)
# empty = []
# empty.pop()  # IndexError: pop from empty list
```

## 6. `clear()` Method

**Description:** Removes all elements from the list, making it empty.

**Syntax:** `list.clear()`

**Returns:** None (modifies the list in-place)

```python
# Example 1: Clear fruits list
fruits = ['apple', 'banana', 'orange']
fruits.clear()
print(fruits)  # Output: []

# Example 2: Clear numbers list
numbers = [1, 2, 3, 4, 5]
numbers.clear()
print(numbers)  # Output: []
print(f"Length: {len(numbers)}")  # Output: Length: 0
```

## 7. `index()` Method

**Description:** Returns the index of the first occurrence of a specified element. Optional start and end parameters can limit the search range.

**Syntax:** `list.index(element, start=0, end=len(list))`

**Returns:** Integer (index of the element)

**Raises:** ValueError if element is not found

```python
# Example 1: Find index of element
fruits = ['apple', 'banana', 'orange', 'banana']
idx = fruits.index('banana')
print(f"Index of banana: {idx}")  # Output: Index of banana: 1

# Example 2: Find with duplicates
numbers = [10, 20, 30, 20, 40]
idx = numbers.index(20)
print(f"Index of 20: {idx}")  # Output: Index of 20: 1

# Example 3: Search in a specific range
idx = numbers.index(20, 2)  # Start from index 2
print(f"Index of 20 (from index 2): {idx}")  # Output: Index of 20 (from index 2): 3

# Example 4: Error when not found (commented to avoid error)
# fruits.index('grape')  # ValueError: 'grape' is not in list
```

## 8. `count()` Method

**Description:** Returns the number of times a specified element appears in the list.

**Syntax:** `list.count(element)`

**Returns:** Integer (count of occurrences)

```python
# Example 1: Count occurrences in fruits
fruits = ['apple', 'banana', 'orange', 'banana', 'banana']
count = fruits.count('banana')
print(f"Count of banana: {count}")  # Output: Count of banana: 3

# Example 2: Count in numbers
numbers = [1, 2, 3, 2, 4, 2, 5]
count = numbers.count(2)
print(f"Count of 2: {count}")  # Output: Count of 2: 3

# Example 3: Element not in list returns 0
count = fruits.count('grape')
print(f"Count of grape: {count}")  # Output: Count of grape: 0
```

## 9. `sort()` Method

**Description:** Sorts the list in ascending order by default. Can be customized with key and reverse parameters.

**Syntax:** `list.sort(key=None, reverse=False)`

**Returns:** None (modifies the list in-place)

```python
# Example 1: Sort numbers in ascending order
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()
print(f"Sorted ascending: {numbers}")  # Output: Sorted ascending: [1, 1, 2, 3, 4, 5, 9]

# Example 2: Sort in descending order
numbers2 = [3, 1, 4, 1, 5, 9, 2]
numbers2.sort(reverse=True)
print(f"Sorted descending: {numbers2}")  # Output: Sorted descending: [9, 5, 4, 3, 2, 1, 1]

# Example 3: Sort strings alphabetically
fruits = ['orange', 'apple', 'banana']
fruits.sort()
print(f"Sorted fruits: {fruits}")  # Output: Sorted fruits: ['apple', 'banana', 'orange']

# Example 4: Sort with key function (by length)
words = ['apple', 'pie', 'zoo', 'at']
words.sort(key=len)
print(f"Sorted by length: {words}")  # Output: Sorted by length: ['at', 'pie', 'zoo', 'apple']
```

## 10. `reverse()` Method

**Description:** Reverses the order of elements in the list.

**Syntax:** `list.reverse()`

**Returns:** None (modifies the list in-place)

```python
# Example 1: Reverse fruits list
fruits = ['apple', 'banana', 'orange']
fruits.reverse()
print(f"Reversed fruits: {fruits}")  # Output: Reversed fruits: ['orange', 'banana', 'apple']

# Example 2: Reverse numbers
numbers = [1, 2, 3, 4, 5]
numbers.reverse()
print(f"Reversed numbers: {numbers}")  # Output: Reversed numbers: [5, 4, 3, 2, 1]

# Example 3: Reversing an empty list
empty = []
empty.reverse()
print(f"Reversed empty: {empty}")  # Output: Reversed empty: []
```

## 11. `copy()` Method

**Description:** Returns a shallow copy of the list. Changes to the copy don't affect the original list (for simple data types).

**Syntax:** `list.copy()`

**Returns:** A new list (shallow copy of the original)

```python
# Example 1: Copy and modify independently
fruits = ['apple', 'banana', 'orange']
fruits_copy = fruits.copy()
fruits_copy.append('mango')

print(f"Original: {fruits}")       # Output: Original: ['apple', 'banana', 'orange']
print(f"Copy: {fruits_copy}")  # Output: Copy: ['apple', 'banana', 'orange', 'mango']

# Example 2: Copy numbers and modify
numbers = [1, 2, 3]
numbers_copy = numbers.copy()
numbers_copy[0] = 99

print(f"Original: {numbers}")       # Output: Original: [1, 2, 3]
print(f"Copy: {numbers_copy}")  # Output: Copy: [99, 2, 3]
```

## Quick Reference Table

| Method | Description | Returns | Modifies List |
|--------|-------------|---------|---------------|
| `append(x)` | Add item to end | None | âœ“ |
| `extend(iter)` | Add all items from iterable | None | âœ“ |
| `insert(i, x)` | Insert item at position | None | âœ“ |
| `remove(x)` | Remove first occurrence | None | âœ“ |
| `pop(i)` | Remove and return item at index | Item | âœ“ |
| `clear()` | Remove all items | None | âœ“ |

```python
# Practice: All methods together
my_list = [1, 2, 3]
print(f"Original: {my_list}")

my_list.append(4)
print(f"After append(4): {my_list}")

my_list.extend([5, 6])
print(f"After extend([5, 6]): {my_list}")

my_list.insert(0, 0)
print(f"After insert(0, 0): {my_list}")

my_list.remove(3)
print(f"After remove(3): {my_list}")

popped = my_list.pop()
print(f"After pop(): {my_list}, popped value: {popped}")

print(f"Index of 2: {my_list.index(2)}")
print(f"Count of 2: {my_list.count(2)}")

my_list_copy = my_list.copy()
print(f"Copy: {my_list_copy}")

my_list.sort()
print(f"After sort(): {my_list}")

my_list.reverse()
print(f"After reverse(): {my_list}")
```

## Tips & Best Practices

### ðŸ’¡ append() vs extend()
Use `append()` to add a single element, `extend()` to add multiple elements from an iterable.

### âš ï¸ In-place vs New List
Most list methods modify the list in-place and return `None`. Use `sorted()` and `reversed()` built-in functions if you need new lists.

### ðŸ“‹ Shallow Copy
The `copy()` method creates a shallow copy. For nested lists, use `copy.deepcopy()` from the copy module.

### âš¡ remove() and index()
These methods raise `ValueError` if the element is not found. Use `in` operator to check existence first if needed.

**Question 5:** Create a function that takes a list and rotates it to the right by n positions. For example, `rotate([1, 2, 3, 4, 5], 2)` should return `[4, 5, 1, 2, 3]`. Use list methods like pop() and insert().

```python
# Write your answer here
def rotate(lst, n):
    for i in range(n):
        lst.insert(0,lst.pop())
    return lst

print(rotate([1, 2, 3, 4, 5], 2))
```
